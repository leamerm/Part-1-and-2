```python
## Arrival time function for StopPoint
def get_arrival_times(stop_id):
    url = f"https://api.tfl.gov.uk/StopPoint/{stop_id}/Arrivals?app_id={app_id}&app_key={app_key}"
    print('loading:',url)
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None

## StopPoint IDs for Holborn Station and nearby bus stops
holborn_station_id = '940GZZLUHBN'
bus_stop_ids = [
    '490000091A',
    '490000091B'
]

## Arrival times for Holborn Station
holborn_arrivals = get_arrival_times(holborn_station_id)

## Arrival times for bus stations
bus_stop_arrivals = [];
for bus_stop_id in bus_stop_ids:
    bus_stop_arrivals.append(get_arrival_times(bus_stop_id))
```

## Build time table and countdown:

```python
timetable = [];

# build timetable for Holborn Underground Station:
for elem in holborn_arrivals:
    stationName = elem['stationName'];
    expectedArrival = elem['expectedArrival'];
    expectedArrival_dt = datetime.strptime(elem['expectedArrival'], '%Y-%m-%dT%H:%M:%SZ');
    expectedArrival_minutes = (datetime.now() - expectedArrival_dt).seconds/60
    destinationName = elem['destinationName'];
    direction = elem['direction'];
    # arrival soon?
    if direction=='inbound': 
        timetable.append([stationName,destinationName,expectedArrival,expectedArrival_minutes]);

# build timetable for bus stations:
#for bus_stop_arrival in bus_stop_arrivals:
    # parse in the same way than u station above and add elements to timetable
    
```
